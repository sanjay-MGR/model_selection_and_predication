```python
#import laiberis
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load datasets
data = sns.load_dataset('mpg')
```

```python
data.head(3)
```

```python
#import the model(Rgression)
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
```

```python
#build the model
linear = LinearRegression()
tree = DecisionTreeRegressor()
svm = SVR()
forest = RandomForestRegressor()
```

```python
#split data into x and y
df = data[['mpg','cylinders','displacement','horsepower','weight','acceleration']]
```

```python
df.isna().sum()
```

```python
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean())
```

```python
df.isna().sum()
```

```python
x = df.drop('mpg',axis = 1)
```

```python
y = df['mpg']
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
#fit the model
linear.fit(x_train,y_train)
```

```python
tree.fit(x_train,y_train)
```

```python
svm.fit(x_train,y_train)
```

```python
forest.fit(x_train,y_train)
```

```python
#predict the value
linear_test = linear.predict(x_test)
liner_train = linear.predict(x_train)
```

```python
tree_train = tree.predict(x_train)
tree_test = tree.predict(x_test)
```

```python
forest_train = forest.predict(x_train)
forest_test = forest.predict(x_test)
```

```python
svm_train = svm.predict(x_train)
svm_test = svm.predict(x_test)
```

```python
#check th accuracy
from sklearn.metrics import r2_score
```

```python
#linear regression
r2_train_linear = r2_score(y_train,liner_train)
r2_train_linear
```

```python
r2_test_linear = r2_score(y_test,linear_test)
r2_test_linear
```

```python
#decion tree regression
r2_tree_train = r2_score(y_train,tree_train)
r2_tree_train
```

```python
r2_tree_test = r2_score(y_test,tree_test)
r2_tree_test
```

```python
#random forest Regressor
r2_forest_train = r2_score(y_train,forest_train)
r2_forest_train
```

```python
r2_forest_test = r2_score(y_test,forest_test)
r2_forest_test
```

```python
#svm
r2_svm_train = r2_score(y_train,svm_train)
r2_svm_train
```

```python
r2_svm_test = r2_score(y_test,svm_test)
r2_svm_test
```

```python
#after analizing the above model
#the best accuracy score is get from random forest Regressor model
#so we select the random forest Recursion model to get the output
```
